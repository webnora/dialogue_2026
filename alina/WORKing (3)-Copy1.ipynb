```python
import requests
import pandas as pd
import time
```

```python
API_KEY = "395dd2df-7520-4b46-8ba7-6d4077fa9c2a"
BASE_URL = "http://content.guardianapis.com/search"
MAX_PAGES = 50 
```

```python
# Конфигурация для 4 типов статей
article_configs = [
    {"type": "Analytical", "tag": "tone/analysis"},
    {"type": "Feature", "tag": "tone/features"},
    {"type": "Editorial", "tag": "tone/editorials"},
    {"type": "Review", "tag": "tone/reviews"}
    {"type": "Review", "tag": "tone/news"}
]
```

```python
all_articles = []

for config in article_configs:
    print(f"Собираем {config['type']}...")
    current_page = 1
    
    while current_page <= MAX_PAGES:
        params = {
            "api-key": API_KEY,
            "tag": config["tag"],
            "page": current_page,
            "page-size": 200,
            "show-fields": "body,headline,byline,publication,trailText"
        }
        
        try:
            response = requests.get(BASE_URL, params=params)
            data = response.json()
            articles = data["response"]["results"]
            
            if not articles:
                break  # Больше нет статей
                
            for article in articles:
                article["category"] = config["type"]
                all_articles.append(article)
            
            print(f"Страница {current_page} собрана")
            current_page += 1
            time.sleep(0.5)  # Защита от rate limits
            
        except Exception as e:
            print(f"Ошибка: {e}")
            break
```

```python
# Сохраняем промежуточные данные
pd.DataFrame(all_articles).to_csv("guardian_non_news.csv", index=False)
```
