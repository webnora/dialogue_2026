```python
import pandas as pd
from ast import literal_eval
from bs4 import BeautifulSoup
import re
from tqdm import tqdm

# Инициализация прогресс-бара для pandas
tqdm.pandas()
```

```python
def clean_fields(field_str):
    try:
        # Преобразование строки в словарь
        field_dict = literal_eval(field_str)
        
        # Извлечение текста статьи из HTML
        soup = BeautifulSoup(field_dict["body"], "html.parser")
        text = soup.get_text(separator=" ")
        
        # Удаление ссылок и специальных символов
        text = re.sub(r"http\S+|www\S+|https\S+", "", text)  # Удаление URL
        text = re.sub(r"[^a-zA-Z0-9\s]", "", text)            # Удаление всех символов, кроме букв, цифр и пробелов
        text = re.sub(r"\s+", " ", text).strip().lower()      # Нормализация пробелов
        
        return text
    except:
        return ""  # Возвращаем пустую строку в случае ошибки
```

```python
# Загрузка данных
df = pd.read_csv("combined_guardian.csv")

# Очистка колонки с прогресс-баром
df["cleaned_text"] = df["fields"].progress_apply(clean_fields)

# Удаление исходной колонки
df = df.drop("fields", axis=1)
```

```python
df.to_csv("cleaned_combined_guardian.csv", index=False)
```
